```python
import os
from typing import Any
from pyannote.audio import Pipeline
from pyannote.core import Annotation

def run_speaker_diarization(audio_path: str, local_model_path: str) -> Any:
    """
    使用 本地 pyannote.audio 模型进行说话人分轨
    """
    if not os.path.exists(local_model_path):
        raise FileNotFoundError(f"模型路径不存在: {local_model_path}")
    if not os.path.exists(audio_path):
        raise FileNotFoundError(f"音频文件不存在: {audio_path}")

    print(f"正在加载本地模型: {local_model_path} ...")
    pipeline = Pipeline.from_pretrained(local_model_path)

    print("正在进行说话人分轨...")
    diarization_result = pipeline(audio_path)

    return diarization_result

def print_diarization_result(diarization: Any):
    """
    适配社区版模型的打印函数
    """
    print("\n" + "=" * 50)
    print("说话人分轨结果")
    print("=" * 50)
    
    # 核心修复：访问 speaker_diarization 属性
    if hasattr(diarization, 'speaker_diarization'):
        annotation = diarization.speaker_diarization
        # 使用原有逻辑遍历 Annotation 对象
        for segment, _, speaker in annotation.itertracks(yield_label=True):
            print(f"[{segment.start:.2f}s - {segment.end:.2f}s] {speaker}")
    # 兼容旧版直接返回 Annotation 的情况
    elif isinstance(diarization, Annotation):
        for segment, _, speaker in diarization.itertracks(yield_label=True):
            print(f"[{segment.start:.2f}s - {segment.end:.2f}s] {speaker}")
    else:
        print("无法识别的分轨结果格式！")
        print(f"结果类型: {type(diarization)}")

if __name__ == "__main__":
    # ------------------ 配置区域 ------------------
    YOUR_AUDIO_PATH = "/home/user512-003/songcw/data/huilongguan_zisha/201501151401/R000331.wav"
    LOCAL_MODEL_DIR = "/home/user512-003/songcw/Pretrained_Model/pyannote/speaker-diarization-community-1"
    # ---------------------------------------------

    # 运行分轨
    diarization_result = run_speaker_diarization(YOUR_AUDIO_PATH, LOCAL_MODEL_DIR)

    # 打印结果
    print_diarization_result(diarization_result)
```
